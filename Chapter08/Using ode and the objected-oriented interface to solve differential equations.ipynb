```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
```

```python
# differential equation is (3x-1)y''-(3x+2)y'+(6x-8)y=0, y(0)=2, y'(0)=3  -AK
def g(y, t):
    y0 = y[0]
    y1 = y[1]
    y2 = ((3*t+2)*y1 + (6*t-8)*y0)/(3*t-1)
    return y1, y2
```

```python
# Initial conditions on y, y' at x=0
init = 1, 0
# First integrate from 0 to 2
t = np.linspace(0,2,100)
sol=odeint(g, init, t)
```

```python
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
